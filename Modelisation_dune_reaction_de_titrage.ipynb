# **Evolution des quantités de matière lors d'un titrage**

```python
#Grandeur du problème
C = 0.025 #mmol/mL
Vo = 20 #mL
Co = 0.00738 #mmol/mL
Veq = 5.9 #mL


def Avant_Eq(V):
  no_v=Co*Vo-C*V
  n_v=0
  np_v=C*V #produit
  nsp_v=C*V #ions spectateurs
  return [no_v,n_v,np_v,nsp_v]

#Fonction calculant la composition du système après l'équivalence (incluse)
def Apres_Eq(V):
  no_v=0
  n_v=C*V-C*Veq
  np_v=C*Veq
  nsp_v=C*V
  return [no_v,n_v,np_v,nsp_v]


#Définition du domaine des abscisses V volume versé de 0 à 2*VEq
V = [i*Veq/10 for i in range(21)] #en L

#Initialisation des listes vides
no, n, np, nsp = [], [], [], []

#pour chaque valeur x du volume V de solution titrante versé
for x in V:
  if x<=Veq: #Avant l'quivalence
    no.append(Avant_Eq(x)[0])
    n.append(Avant_Eq(x)[1])
    np.append(Avant_Eq(x)[2])
    nsp.append(Avant_Eq(x)[3])
  else: #Après l'quivalence
    no.append(Apres_Eq(x)[0])
    n.append(Apres_Eq(x)[1])
    np.append(Apres_Eq(x)[2])
    nsp.append(Apres_Eq(x)[3])
  
#Tracé des courbes
plt.figure(figsize=(10,7))
plt.plot(V,no, label='n($Cl^{-}$)=$n_{titré}$')
plt.plot(V,n, label='n($Ag^{+}$)=$n_{titrant}$')
plt.plot(V,np, label='n($AgCl_{(s)}$)')
plt.plot(V,nsp,':', label='n($NO_{3}^{-}$)')

plt.xlabel('Volume $V_{versé}$ de solution titrante $AgNO_{3}$ (en mL)')
plt.ylabel('Quantité de matière (en mol)')
plt.title('Evolution des quantités de matière des espèces lors du titrage de l eau d une bouteille de Saint-Yorre')
plt.plot([5.9, 5.9], [0, 0.25],'r',linewidth=3, linestyle='--', label="Veq")
plt.grid(linestyle='--')
plt.xlim(0,11.8)
plt.ylim(0,0.40)
plt.legend(loc=2, prop={'size': 15})
plt.show()
```

# **Evolution de la conductivité lors d'un titrage**

```python
#W = np.array(V)
#No = np.array(no)
#N = np.array(n)
#Np = np.array(np)
#Nsp = np.array(nsp)

sigma=[]
for i in range(len(V)):
  sigma.append((no[i]*7600+n[i]*6200+nsp[i]*7100)/(Vo+150+V[i]))

plt.figure(figsize=(16,13))

plt.subplot(2,2,3)

plt.xlabel('Volume $V_{versé}$ de solution titrante $AgNO_{3}$ (en mL)')
plt.ylabel('$\sigma$ (en mS.$m^{-1}$)')
plt.title('Suivi conductimétrique du titrage')
plt.grid(linestyle='--')
plt.plot(V,sigma,linewidth=3, label='$\sigma$')
plt.xlim(0,11.8)
plt.ylim(0,25)
plt.legend(loc=2, prop={'size': 20})


plt.subplot(2,2,4)

plt.xlabel('Volume $V_{versé}$ de solution titrante $AgNO_{3}$ (en mL)')
plt.ylabel('$\sigma$ (en mS.$m^{-1}$)')
plt.title('Suivi conductimétrique du titrage')
plt.grid(linestyle='--')
plt.plot(V,sigma,linewidth=3, label='$\sigma$')
plt.plot([5.9, 5.9], [0, 15],'r',linewidth=3, linestyle='--', label="Veq")
plt.xlim(0,11.8)
plt.ylim(0,25)
plt.legend(loc=2, prop={'size': 20})

plt.subplot(2,2,1)

plt.xlabel('Volume $V_{versé}$ de solution titrante $AgNO_{3}$ (en mL)')
plt.ylabel('$\sigma$ (en mS.$m^{-1}$)')
plt.title('Suivi conductimétrique du titrage')
plt.grid(linestyle='--')
plt.plot([0, 5.9], [7.9, 7.1],'g',linewidth=3, linestyle='--', label="$\sigma \propto (\lambda_{NO_{3}^{-}} - \lambda_{Cl^{-}})*V_{versé}$")
plt.xlim(0,11.8)
plt.ylim(0,25)
plt.legend(loc=2, prop={'size': 20})

plt.subplot(2,2,2)

plt.xlabel('Volume $V_{versé}$ de solution titrante $AgNO_{3}$ (en mL)')
plt.ylabel('$\sigma$ (en mS.$m^{-1}$)')
plt.title('Suivi conductimétrique du titrage')
plt.grid(linestyle='--')
plt.plot([5.9, 14], [5.9, 19.4],'y',linewidth=3, linestyle='--', label="$\sigma \propto (\lambda_{NO_{3}^{-}} + \lambda_{Ag^{+}})*V_{versé}$")
plt.xlim(0,11.8)
plt.ylim(0,25)
plt.legend(loc=2, prop={'size': 20})


plt.show()
```
