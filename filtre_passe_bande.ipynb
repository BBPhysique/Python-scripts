```python
#Une question  propos de ce script ? --> brian.bernard@ens-paris-saclay.fr

%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np
import numpy.random as rd




def f(Q,f0):
  fig = plt.figure(figsize=(13,5))
  plt.rcParams.update({'font.size': 18})
  ax = fig.add_subplot(1, 1, 1)

  #print(sigmafreq/freq)
  #sigma l'cart-type en nm de la gaussienne du spectre centr sur 633 nm
  x = np.logspace(0, 7, num=1000)

  y = 1/(1+(Q**2)*(x/f0-f0/x)**2)**0.5
  #ax.set_yscale('log')
  ax.set_xscale('log')
  #calcul et trac de la proportion en base
  ax.plot(x, 20*np.log10(y) ,color = 'red') 
  #calcul et trac de la proportion en acide
  #plt.plot(x, 100/(1+10**(x-pKa)),color = 'blue', label = 'acide') 
  
  ax.axvline(x=f0, linestyle='dashed', color='black',label = 'f0')
  ax.axhline(y=0, linestyle='dashed', color='black',label = 'G = 0')

  xgauche = np.logspace(0, np.log10(f0), num=1000)
  ax.plot(xgauche,20*np.log10(1/((Q**2)*(f0/xgauche)**2)**0.5), linestyle='dashed', color='orange',label = 'asymptotes')
  xdroite = np.logspace(np.log10(f0),7, num=1000)
  ax.plot(xdroite,20*np.log10(1/((Q**2)*(xdroite/f0)**2)**0.5), linestyle='dashed', color='orange')


  #mise en forme du graphique
  #plt.ylim(-0.05*100, 1.05*100)
  plt.ylabel("Gain $ G = 20log( | H(f) | ) $")
  plt.xlabel("frquence $f$ (Hz)")
  plt.title("Diagramme de Bode, Q = {}, f0 = {} Hz".format(Q,f0))
  plt.legend()
  #plt.legend()
  plt.show()
             
#actualisation de la barre d'actualisation dynamique
interactive_plot = interactive(f, Q=(0.1, 10,1),f0 = (1000,5000,1000))
output = interactive_plot.children[-1]
output.layout.height = '350px'
interactive_plot
```

```python
fig = plt.figure(figsize=(13,8))
plt.rcParams.update({'font.size': 18})
ax = fig.add_subplot(1, 1, 1)

#print(sigmafreq/freq)
#sigma l'cart-type en nm de la gaussienne du spectre centr sur 633 nm
x = np.logspace(0, 7, num=1000)

f0 = 3000
colorlist = ['red','green','blue','orange']
Qlist = [0.1,1,10,100]
for Qindex in range(len(Qlist)) : 
  Q = Qlist[Qindex]
  couleur = colorlist[Qindex]
  y = 1/(1+(Q**2)*(x/f0-f0/x)**2)**0.5
  #ax.set_yscale('log')
  ax.set_xscale('log')
  #calcul et trac de la proportion en base
  ax.plot(x, 20*np.log10(y) ,color = couleur,label = "Q = {}".format(Q)) 
  #calcul et trac de la proportion en acide
  #plt.plot(x, 100/(1+10**(x-pKa)),color = 'blue', label = 'acide') 

  xgauche = np.logspace(0, np.log10(f0), num=1000)
  ax.plot(xgauche,20*np.log10(1/((Q**2)*(f0/xgauche)**2)**0.5), linestyle='dashed', color=couleur)
  xdroite = np.logspace(np.log10(f0),7, num=1000)
  ax.plot(xdroite,20*np.log10(1/((Q**2)*(xdroite/f0)**2)**0.5), linestyle='dashed', color=couleur)

ax.axvline(x=f0, linestyle='dashed', color='black',label = 'f0')
ax.axhline(y=0, linestyle='dashed', color='black',label = 'G = 0')

plt.ylabel("Gain $ G = 20log( | H(f) | ) $")
plt.xlabel("frquence $f$ (Hz)")
plt.title("Diagramme de Bode f0 = {} Hz".format(f0))
plt.legend()
#plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(13,8))
plt.rcParams.update({'font.size': 18})
ax = fig.add_subplot(1, 1, 1)

fmin = 3.2
fmax = 3.8

#print(sigmafreq/freq)
#sigma l'cart-type en nm de la gaussienne du spectre centr sur 633 nm
x = np.logspace(fmin, fmax, num=1000)

f0 = 3000
colorlist = ['green','orange']
Qlist = [5,20]
for Qindex in range(len(Qlist)) : 
  Q = Qlist[Qindex]
  couleur = colorlist[Qindex]
  y = 1/(1+(Q**2)*(x/f0-f0/x)**2)**0.5
  #ax.set_yscale('log')
  ax.set_xscale('log')
  #calcul et trac de la proportion en base
  ax.plot(x, 20*np.log10(y) ,color = couleur,label = "Q = {}".format(Q)) 
  #calcul et trac de la proportion en acide
  #plt.plot(x, 100/(1+10**(x-pKa)),color = 'blue', label = 'acide') 

  xgauche = np.logspace(fmin, np.log10(f0), num=1000)
  ax.plot(xgauche,20*np.log10(1/((Q**2)*(f0/xgauche)**2)**0.5), linestyle='dotted', color=couleur)
  xdroite = np.logspace(np.log10(f0),fmax, num=1000)
  ax.plot(xdroite,20*np.log10(1/((Q**2)*(xdroite/f0)**2)**0.5), linestyle='dotted', color=couleur)
  ax.axvline(x=f0-f0/(2*Q),ymax=0.87, linestyle='dashed', color=couleur)
  ax.axvline(x=f0+f0/(2*Q),ymax=0.87, linestyle='dashed', color=couleur)

ax.axvline(x=f0,linestyle='dashed', color='black',label = 'f0')
ax.axhline(y=-3, linestyle='dashed', color='black',label = 'G = -3dB')

plt.ylabel("Gain $ G = 20log( | H(f) | ) $")
plt.xlabel("frquence $f$ (Hz)")
plt.title("Diagramme de Bode f0 = {} Hz".format(f0))
plt.legend()
#plt.legend()
plt.show()
```
