```python
#Une question  propos de ce script ? --> brian.bernard@ens-paris-saclay.fr

%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np
import numpy.random as rd



c = 3*10**8
longonde = 633*10**-9
freq = c/longonde

nbperiodes = 100
nbptsperiode = 50

def f(N,sigma):
  plt.figure(figsize=(13,5))
  plt.rcParams.update({'font.size': 18})

  sigma = sigma*10**-9 #retour en S.I.
  sigmafreq = sigma*c/(longonde**2)
  #print(sigmafreq/freq)
  #sigma l'cart-type en nm de la gaussienne du spectre centr sur 633 nm
  x = np.linspace(-nbperiodes/(2*freq), nbperiodes*1/(2*freq), num=nbperiodes*nbptsperiode)
  y = 0
  for onde in range(N) :
    randfreq = rd.normal(freq, sigmafreq, 1)
    y = y+np.cos(2*np.pi*x*randfreq + rd.uniform(-np.pi,np.pi))
  
  #calcul et trac de la proportion en base
  plt.plot(x, y ,color = 'red') 
  #calcul et trac de la proportion en acide
  #plt.plot(x, 100/(1+10**(x-pKa)),color = 'blue', label = 'acide') 
  
  #mise en forme du graphique
  #plt.ylim(-0.05*100, 1.05*100)
  plt.ylabel("Intensit normalise $\\frac{I}{I_0}$")
  plt.xlabel("temps t")
  plt.title("Somme de {} ondes  633 nm".format(round(N,2)))
  
  #plt.legend()
  plt.show()
             
#actualisation de la barre d'actualisation dynamique
interactive_plot = interactive(f, N=(1, 1000,100),sigma = (1,10,1))
output = interactive_plot.children[-1]
output.layout.height = '350px'
interactive_plot
```

- dsormais la phase  l'origine de chaque photon est alatoire

- $\sigma$ est l'cart-type en nm de la gaussienne modlisant le spectre de la source

- Origine des phases commune  t = 0 ($\sim$ instant de sortie du laser) 

- Intensit d'un photon prise gale  $I_0$
