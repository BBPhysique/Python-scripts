```python
import random as rd
import numpy as np
from math import *
import matplotlib.pyplot as plt
```

```python
N=500
A=np.linspace(5,200,195)

B=np.zeros(500)

```

```python
# energie volumique
a_v=15.6
E_v=a_v*A
B=E_v

plt.figure(figsize=(8,5))
plt.plot(A,E_v/A,'b',label='terme volumique')
plt.xlim(30,200)
plt.ylim(0,20)
plt.title("Modèle de la goutte liquide de Bethe-Weizsäcker")
plt.xlabel('A')
plt.ylabel('B/A (MeV)')
plt.legend(bbox_to_anchor=(1.0, 1.0),loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#energie de surface
a_s=17.2
E_s=-a_s*A**(2/3)
B=E_v+E_s

plt.figure(figsize=(8,5))
plt.plot(A,(E_v+E_s)/A,'g',label='terme surfacique')
plt.plot(A,E_v/A,'b',label='terme volumique')
plt.xlim(30,200)
plt.ylim(0,20)
plt.title("Modèle de la goutte liquide de Bethe-Weizsäcker")
plt.xlabel('A')
plt.ylabel('B/A (MeV)')
plt.legend(bbox_to_anchor=(1.0, 1.0),loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#energie electrostatique
a_c=0.7
Z=30
E_c=-a_c*(Z**2)/(A**(1/3))
B=E_v+E_s+E_c

plt.figure(figsize=(8,5))
plt.plot(A,(E_v+E_s+E_c)/A,'r',label='terme coulombien')
plt.plot(A,(E_v+E_s)/A,'g',label='terme surfacique')
plt.plot(A,E_v/A,'b',label='terme volumique')
plt.xlim(30,200)
plt.ylim(0,20)
plt.title("Modèle de la goutte liquide de Bethe-Weizsäcker")
plt.xlabel('A')
plt.ylabel('B/A (MeV)')
plt.legend(bbox_to_anchor=(1.0, 1.0),loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#energie d'asymétrie
a_a=23.6
E_a=-a_a*((A-2*Z)**2)/A
B=B+E_a

plt.figure(figsize=(8,5))
plt.plot(A,(E_v+E_s+E_c)/A,'r',label='terme coulombien')
plt.plot(A,(E_v+E_s)/A,'g',label='terme surfacique')
plt.plot(A,E_v/A,'b',label='terme volumique')
plt.plot(A,(E_v+E_s+E_c+E_a)/A,'orange',label='terme d asymétrie')
plt.xlim(30,200)
plt.ylim(0,20)
plt.title("Modèle de la goutte liquide de Bethe-Weizsäcker")
plt.xlabel('A')
plt.ylabel('B/A (MeV)')
plt.legend(bbox_to_anchor=(1.0, 1.0),loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#energie d'appariement
a_A=11.2
E_A=a_A/np.sqrt(A)
B=B+E_A

plt.figure(figsize=(8,5))
plt.plot(A,(E_v+E_s+E_c)/A,'r',label='terme coulombien')
plt.plot(A,(E_v+E_s)/A,'g',label='terme surfacique')
plt.plot(A,E_v/A,'b',label='terme volumique')
plt.plot(A,(E_v+E_s+E_c+E_a)/A,'orange',label='terme d asymétrie')
plt.plot(A,(E_v+E_s+E_c+E_a+E_A)/A,'purple',label='terme d appariement')
plt.xlim(30,200)
plt.ylim(0,20)
plt.title("Modèle de la goutte liquide de Bethe-Weizsäcker")
plt.xlabel('A')
plt.ylabel('B/A (MeV)')
plt.legend(bbox_to_anchor=(1.0, 1.0),loc='upper left')
plt.tight_layout()
plt.show()
```

Le terme d'appariement change rien donc pas très intéressant. Dans le Aslangul, il ne le met qu'en remarque donc peut être pas mal de ne pas forcément en parler


```python
#isobare le plus stable

Z1=np.linspace(20,100,200)
A1=50
B1=a_v*A1-a_s*A1**(2/3)-a_c*(Z1**2)/(A1**(1/3))-a_a*((A1-2*Z1)**2)/A1

maxi = B1[0]
longueur=len(B1)
indice_max = 0
for i in range(longueur):
    if B1[i] >= maxi:
        maxi = B1[i]
        indice_max = i
        
print('Valeur de Z qui rend B maximal : Z =',Z1[indice_max])

plt.figure(figsize=(8,5))
plt.plot(Z1,B1,'b',label='Energie de liaison')
plt.plot(Z1,-a_c*(Z1**2)/(A1**(1/3)),'r',label='répulsion coulombienne')
plt.plot(Z1,a_v*A1-a_s*A1**(2/3)-a_a*((A1-2*Z1)**2)/A1,'g',label='asymétrie')
plt.xlim(20,60)
plt.ylim(-1000,1000)
plt.xlabel('Z')
plt.ylabel('B(A,Z)')
plt.title('Energie de liaison en fonction de Z pour A fixé')
plt.legend(loc='upper right')
plt.show()
```
