```python
#Une question à propos de ce script ? --> brian.bernard@ens-paris-saclay.fr

%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np
#from IPython.display import Javascript
#display(Javascript('''google.colab.output.setIframeHeight(0, true, {maxHeight: 50000})'''))
import warnings; warnings.simplefilter('ignore')
from IPython.display import Javascript
def resize_colab_cell():
  display(Javascript('google.colab.output.setIframeHeight(0, true, {maxHeight: 5000})'))
get_ipython().events.register('pre_run_cell', resize_colab_cell)

plt.rcParams.update({'font.size': 18})

c = 3*10**8
#définition de l'intervalle de points de calcul

omegalaser = 2*np.pi*c/(1.31*10**-6)
omega = np.linspace(10**13, omegalaser, num=1000)
#calcul et tracé de la proportion en base
def f(a):
    plt.figure(figsize=(13,5))
    a = a*10**-6
    
    for p in range(7):
      plt.plot(omega, np.sqrt((omega/c)**2-np.pi**2*p**2/a**2), label = "p = {}".format(p)) 
    #calcul et tracé de la proportion en acide
    #plt.plot(x, 100/(1+10**(x-pKa)),color = 'blue', label = 'acide') 

    #mise en forme du graphique
    #plt.ylim(-0.05*100, 1.05*100)
    plt.ylabel("vecteur d'onde selon z $k_z$ ($m^{-1})$")
    plt.xlabel("pulsation $\\omega$ (rad/s)")
    plt.title("diamètre de la fibre : {} µm".format(round(a*10**6),2))
    plt.axvline(x=omegalaser, linestyle='dashed', color='black',label = 'pulsation injectée')
    plt.legend()
    plt.show()


        #actualisation de la barre d'actualisation dynamique
interactive_plot = interactive(f, a=(10,60,10))
output = interactive_plot.children[-1]
output.layout.height = '350px'
interactive_plot  
```

```python
plt.figure(figsize=(13,5))
for r in [0.6,0.8,0.95,0.99]:
    finesse = np.pi*r/(1-r**2)
    #définition de l'intervalle de points de calcul
    x = np.linspace(-np.pi, np.pi, num=10000)
    
    #calcul et tracé de la proportion en base
    plt.plot(x, 1/(1+(4*r**2/(1-r**2)**2)*(np.sin(x/2))**2),label = "r = {}".format(r)) 
    #calcul et tracé de la proportion en acide
    #plt.plot(x, 100/(1+10**(x-pKa)),color = 'blue', label = 'acide') 
    
    #mise en forme du graphique
    #plt.ylim(-0.05*100, 1.05*100)
    plt.ylabel("Intensité normalisée $\\frac{I}{I_0}$")
    plt.xlabel("Déphasage $\Delta \phi$")
    plt.legend()
    plt.title("Récapitulatif de l'effet de la réflectivité des miroirs".format(round(finesse,2)))
```
